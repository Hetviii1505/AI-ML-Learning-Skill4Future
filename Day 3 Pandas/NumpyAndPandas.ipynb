### Reindexing 
##### modifies the row and column labels of a DataFrame.
##### it denotes the conforming of DataFrame to new index with optional filling logic.

```python
import pandas as pd
# Create dataframe
info = pd.DataFrame({"P":[4, 7, 1, 8, 9], 
                     "Q":[6, 8, 10, 15, 11], 
                     "R":[17, 13, 12, 16, 14], 
                     "S":[15, 19, 7, 21, 9]}, 
                    index =["Parker", "William", "Smith", "Terry", "Phill"])
#Print dataframe
info
```

```python

# reindexing with new index values 
info.reindex(["A", "B", "C", "D", "E"])

```

```python
# Fill missing values
info.reindex (["A", "B", "C", "D", "E"], fill_value =100)
```

### Pandas Sorting 
-- By label 
-- By actual value 

```python
unsorted_df = pd.DataFrame(np.random.randn (10,2), index= [1,4,6,2,3,5,9,8,8,7], columns = ['co12', 'col1'])
sorted_df = unsorted_df.sort_index() 
print (sorted_df)
```

### Sort The Column

```python
# By value 
sorted_df = unsorted_df.sort_values(by='col1')
print (sorted_df)
```
