```python

data = pd.read_csv('creditcard.csv')
print(data.head())
print(data.info())
print(data.describe())
print(data.isnull().sum())
```

```python
plt.hist(data['Time'], bins=30, color='blue', alpha=0.7)
plt.title('Distribution of Transaction Time')
plt.show()
```

```python
# Graph insights can be added here based on the visualizations
plt.hist(data['Amount'])
plt.show()
print("Basic Insights:")
print("Min:", data['Amount'].min())
print("Max:", data['Amount'].max())
print("Mean:", data['Amount'].mean())
print("Median:", data['Amount'].median())
print("Std Dev:", data['Amount'].std())

```
